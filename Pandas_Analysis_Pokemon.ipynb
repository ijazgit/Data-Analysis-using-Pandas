### Making changes to Data 

```python
df["Total"]=df["HP"]+df["Attack"]+df["Defense"]+df["Speed"]+df["Sp. Atk"]+df["Sp. Def"]
```

```python
df.head()
```

```python
df=df.drop(columns=["Total"])  # dropping the column "Total"
```

### Filtering the Data

```python
legend_grass=df.loc[(df["Type 1"]=="Grass")&(df["Type 2"]=="Poison")&(df["HP"]>70)] # for finding grass-poison types with HP greater than 70
legend_grass
```

```python
legend_fire=df.loc[(df["Type 1"]=="Fire")&(df["HP"]>70)]
legend_fire
```
